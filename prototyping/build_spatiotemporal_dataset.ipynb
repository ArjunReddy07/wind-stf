```python
blacklist = ['DE409', 'DE40C', 'DE403',  # outliers in spatial correlogram
             'DE24C', 'DE266', 'DEA2C', ]  # zero installed capacity at 2015-01-01
```

```python
getthis = set( df_spatiotemporal.columns.get_level_values('district') ) - set(blacklist)

df_spatiotemporal.loc[
    :,
    (slice(None), getthis)
]
```

```python
def _get_districts_included(df: pd.DataFrame) -> set:
    return set(df.columns.get_level_values('nuts_id'))

timestamps_to_include = df_spatial.index.intersection(
    df_temporal.index
)

districts_to_include = set( 
    _get_districts_included(df_temporal).intersection( 
        _get_districts_included(df_spatial) )
)

df_spatiotemporal = df_spatiotemporal.loc[
    timestamps_to_include,
    (slice(None), districts_to_include)
]

display(df_spatiotemporal.head(3))
display(df_spatiotemporal.shape[1]/3)
```
