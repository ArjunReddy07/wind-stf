```python
import pandas as pd
import numpy as np
```

```python
power_generated_daily = pd.read_hdf('../data/03_primary/measurements-daily-2000-2015-filtered.hdf', key='measurements_filtered')

display(power_generated_daily.head()); display(power_generated_daily.shape)
```

```python
power_installed_daily = pd.read_hdf('../data/04_feature/power-installed-2000-2015.hdf', key='power_installed')

display(power_installed_daily.head()); display(power_installed_daily.shape)
display(type(power_installed_daily)); display(power_installed_daily.index); display(power_installed_daily.columns)
```

```python
power_installed_daily_trimmed = power_installed_daily.loc[ power_generated_daily.index[0] : power_generated_daily.index[-1] ]

display(power_installed_daily_trimmed.head())
display(power_installed_daily_trimmed.tail()) 
display(power_installed_daily_trimmed.shape) 

display(power_generated_daily.shape)
```

```python
capfactors_mts = power_generated_daily / power_installed_daily_trimmed

display( power_generated_daily.head(3) )
display( power_installed_daily_trimmed.head(3) )
display( capfactors_mts.head(3) )
```
